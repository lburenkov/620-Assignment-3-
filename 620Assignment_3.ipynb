```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
# Set the parameters
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
```

```python
# Read 500 rows
dt = pd.read_excel("network_list.xlsx")
dt
```

```python
dt.shape
```

```python
G = nx.from_pandas_edgelist(dt, source='NETWORK', target='NUMBER_OF_PARTICIPANTS')
nx.draw_networkx(G)
```

```python
print(G)
```

```python
nx.algorithms.diameter(G)
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
nx.draw(G, node_color='bisque', with_labels=True)
```
